```python
import pandas as pd

# 1. 输入/输出文件名（按你的路径改）
infile  = r"E:\wjy\Gravity\OJP_Gravity\data\文献相关地震剖面数据\Jiang,2024,PNG2010_Line01_Moho.txt"
outfile = r"E:\wjy\Gravity\OJP_Gravity\data\文献相关地震剖面数据\PNG2010_Line01_Moho_lonlat.csv"

# 2. 读入原始剖面数据（空格分隔）
df = pd.read_csv(infile, delim_whitespace=True, comment='#')

# 3. 只保留经纬度两列，并改表头为 lon, lat
lonlat = df[['Longitude', 'Latitude']].rename(
    columns={'Longitude': 'lon', 'Latitude': 'lat'}
)

# 4. 写出为 CSV，不带行号
lonlat.to_csv(outfile, index=False, float_format="%.5f")

print(f"Saved {len(lonlat)} points to {outfile}")
```
